```python
import pandas as pd
import numpy as np
import math
#For train Data
dfTrain = pd.read_csv('train.csv')
column_data1 = dfTrain['radius']
column_data2 = dfTrain['height']
y_train = dfTrain['surface_area']
X_train = np.column_stack((column_data1, column_data2))

#For test Data
dfTest = pd.read_csv('test.csv')
column_data3 = dfTest['radius']
column_data4 = dfTest['height']
y_test = dfTest['surface_area']
X_test = np.column_stack((column_data3, column_data4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from joblib import dump

rf_regressor = RandomForestRegressor(n_estimators=120, random_state=0)  #Random Forest Regressor

rf_regressor.fit(X_train, y_train)  #Training

y_pred = rf_regressor.predict(X_test)  #predictions

#Evaluation
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

#dump(rf_regressor, 'model_RFR.pkl')

print(f"Mean Absolute Error: {mae}")
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (R²): {r_squared:.2f}')
```

```python
import matplotlib.pyplot as plt
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.axhline(y=0, color='r', linestyle='--')
```
